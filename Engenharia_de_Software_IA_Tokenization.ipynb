# Introdução à Tokenização

A tokenização é um processo fundamental no campo do Processamento de Linguagem Natural (PLN), onde o texto é transformado em unidades menores chamadas tokens. Esses tokens podem ser palavras individuais, números, pontuações, símbolos especiais ou até mesmo unidades mais complexas, como emojis ou emoticons. ## O que é Tokenização?

Tokenização é o processo de dividir o texto em unidades menores, chamadas tokens. Um token é uma unidade mínima de texto que pode ser tratada como uma única entidade. Por exemplo, em uma frase como "O gato está dormindo.", os tokens podem incluir "O", "gato", "está", "dormindo" e ".". A tokenização é essencial porque ajuda a transformar o texto não estruturado em uma forma que possa ser processada por algoritmos de PLN.




## Por que a Tokenização é Importante?

A tokenização desempenha um papel crítico em várias tarefas de PLN por várias razões:

1. **Preparação de Dados:** Antes que qualquer modelo de PLN possa processar
texto, ele precisa ser tokenizado para que as palavras e outros elementos possam ser tratados individualmente.

2. **Redução de Complexidade:** Ao dividir o texto em tokens, simplificamos o processo de análise, permitindo que os modelos de PLN se concentrem em unidades menores e mais gerenciáveis de texto.

3. **Extração de Informações:** A tokenização permite a extração de informações importantes, como palavras-chave, partes do discurso, entidades nomeadas e muito mais, que são essenciais para muitas tarefas de PLN.

4. **Normalização de Texto:** Em alguns casos, a tokenização também pode envolver a normalização do texto, como a conversão de todas as letras para minúsculas, remoção de pontuações ou símbolos especiais, etc.

## Como a Tokenização é Realizada?

A forma como a tokenização é realizada pode variar dependendo do contexto e das necessidades específicas de cada tarefa de PLN. No entanto, existem algumas abordagens comuns:

1. **Tokenização com Base em Espaços em Branco:** Uma abordagem simples é dividir o texto com base nos espaços em branco. Nesta abordagem, cada palavra é tratada como um token separado.

2. **Tokenização com Base em Regras:** Outra abordagem é usar regras específicas para dividir o texto em tokens. Isso pode envolver o uso de expressões regulares ou outras técnicas para identificar padrões específicos no texto.

3. **Tokenização com Base em Modelos de Linguagem:** Muitas bibliotecas de PLN, como spaCy, NLTK e StanfordNLP, fornecem modelos de linguagem pré-treinados que são capazes de realizar a tokenização de forma mais precisa e sofisticada. Esses modelos são treinados em grandes quantidades de texto e têm conhecimento sobre a estrutura da linguagem, o que lhes permite identificar tokens de forma mais precisa, mesmo em casos complexos, como palavras compostas ou abreviaturas.

# Introdução à Tokenização com spaCy

A tokenização é um processo fundamental em NLP (Processamento de Linguagem Natural), onde o texto é dividido em unidades menores chamadas tokens. Estes tokens podem ser palavras individuais, pontuações, números, ou até mesmo unidades mais complexas em alguns casos. Neste tutorial, vamos explorar os conceitos básicos de tokenização utilizando a biblioteca spaCy em Python.

## Passo 1: Instalação e Importação de Bibliotecas

Antes de começarmos, precisamos instalar o spaCy e carregar o modelo de idioma desejado. No ambiente do Google Colab, podemos fazer isso da seguinte maneira:

Após a instalação, vamos importar as bibliotecas necessárias:

```python
import spacy
import pandas as pd

# Carregando o modelo de idioma em português
nlp = spacy.load('pt_core_news_sm')
```

## Passo 2: Tokenização de Texto

Agora que temos o spaCy configurado, podemos tokenizar um exemplo simples de texto em português:

```python
# Texto de exemplo
texto = "Os alunos da faculdade de Engenharia de Software estão ansiosos para o início do curso!"

# Aplicando a tokenização
doc = nlp(texto)

# Extraindo os tokens
tokens = [token.text for token in doc]
print("Tokens:", tokens)
```

## Passo 3: Análise dos Tokens

Após tokenizar o texto, podemos analisar os tokens para extrair informações adicionais, como posição inicial, se é pontuação, se é espaço em branco, parte do discurso, dependência sintática, etc. Isso pode ser feito da seguinte maneira:

```python
# Coletando informações dos tokens em um único DataFrame
dados_tokens = []
for token in doc:
    dados_tokens.append({
        'Token': token.text,
        'Índice de início': token.idx,
        'É pontuação?': token.is_punct,
        'É espaço em branco?': token.is_space,
        'Parte do discurso': token.pos_,
        'Dependência sintática': token.dep_
    })

# Criando DataFrame com os dados dos tokens
df_tokens = pd.DataFrame(dados_tokens)

# Exibindo o DataFrame
df_tokens
```
